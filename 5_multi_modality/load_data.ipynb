```python
import pandas as pd
import pickle
from PIL import Image
import matplotlib.pyplot as plt
import os

# 
new_dir = "/mnt/data1/zhaoxinjian/zhongkai/TMLR_VISG_/"
os.chdir(new_dir)

#  CSV 
df = pd.read_csv('./modality/PROTEINS/dataset.csv')

# 
n = 5

#  n 
for idx, row in enumerate(df.iterrows()):
    if idx >= n:
        break

    image_path = row[1]['image_path']
    graph_data_path = row[1]['graph_data_path']
    label = row[1]['label']

    # 
    image = Image.open(image_path)

    #  matplotlib 
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.axis('off')
    plt.title(f"Sample {idx + 1}: Label {label}")
    plt.show()

    # 
    with open(graph_data_path, 'rb') as f:
        graph_data = pickle.load(f)

    # 
    print(f"Sample {idx + 1}:")
    print(f"Label: {label}, Graph Data: {graph_data}")
    print("-" * 50)
```
